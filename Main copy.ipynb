```python
movies1 = movies[['id', 'title', 'genre', 'overview']]
movies1
```

```python
movies1['tag']=movies1['overview']+movies1['genre']
```

```python
new_data = movies1.drop(columns=['overview','genre'])
```

```python
new_data
```

1. Bag of word
2. TFIDF


```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python

cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(movies['overview'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarrity = cosine_similarity(vector)
similarrity
```

```python
new_data[new_data['title']=='Once Upon a Time in America'].index[0]
```

```python
distance = sorted(list(enumerate(similarrity[2])),reverse=True,key=lambda vector:vector[1])
for i in distance[0:5]:
  print(movies.iloc[i[0]].title)
```

```python
def recomand(movie):
  index = movies[movies['title']==movie].index[0]
  distance = sorted(list(enumerate(similarrity[index])),reverse=True,key=lambda vector:vector[1])
  for i in distance[0:5]:
    print(movies.iloc[i[0]].title)
```

```python
recomand('Iron Man')
```
